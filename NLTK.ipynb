```python
import pandas as pd
import numpy as np
```

```python
import nltk
nltk.download(['punkt','averaged_perceptron_tagger','maxent_ne_chunker','words'])
```

```python
def get_toked_sent(doc):
    tokenizer = PunktSentenceTokenizer()
    sents = tokenizer.tokenize(doc)
    return sents
```

```python
def get_continuous_chunks(text):
    chunked = ne_chunk(pos_tag(word_tokenize(text)))
    prev = None
    continuous_chunk = []
    current_chunk = []

    for i in chunked:
        if type(i) == Tree:
            current_chunk.append(" ".join([token for token, pos in i.leaves()]))
        elif current_chunk:
            named_entity = " ".join(current_chunk)
            if named_entity not in continuous_chunk:
                continuous_chunk.append(named_entity)
                current_chunk = []
        else:
            continue

    if continuous_chunk:
        named_entity = " ".join(current_chunk)
        if named_entity not in continuous_chunk:
            continuous_chunk.append(named_entity)
    return continuous_chunk[:-1]
```

```python
import glob
import os

data_path = './txt'
text_paths = sorted(glob.glob(os.path.join(data_path,'*.txt')))
docs = []

# chia batch ra, cho đỡ tốn bộ nhớ
for files in text_paths:
    with open(files, mode='r') as f:
        text = f.read()
        docs.append(text)
```

```python
from nltk import ne_chunk, pos_tag, word_tokenize, PunktSentenceTokenizer
from nltk.tree import Tree

nameEnt = []
for doc in docs:
    sents = get_toked_sent(doc)
    for sent in sents:
        entities = get_continuous_chunks(sent)
        nameEnt.extend(entities)
```
